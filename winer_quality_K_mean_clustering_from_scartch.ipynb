#K-Mean algorithm using Python from Scratch

```python
import google.colab
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/winequality.csv')
df.head()
```

```python
plt.scatter(df['quality'],df['fixed acidity'])
plt.xlabel('Quality')
plt.ylabel('Fixed acidity')
plt.title('Wine Data: Quality vs Fixed Acidity ')
```

```python
km = KMeans(n_clusters=2)  # Number of clusters can be changed
y_predicted = km.fit_predict(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
plt.plot(km.cluster_centers_)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
plt.scatter(df1['quality'],df1['fixed acidity'], color='green')
plt.scatter(df2['quality'],df2['fixed acidity'], color='red')

plt.xlabel('Quality')
plt.ylabel('Fixed acidity')
plt.title('Wine Data: Quality vs Fixed Acidity ')
```

```python
plt.scatter(df1['quality'],df1['volatile acidity'], color='green')
plt.scatter(df2['quality'],df2['volatile acidity'], color='red')

plt.xlabel('Quality')
plt.ylabel('Volatile acidity')
plt.title('Wine Data: Quality vs Volatile acidity ')
```

```python
plt.scatter(df1['quality'],df1['pH'], color='green')
plt.scatter(df2['quality'],df2['pH'], color='red')

plt.xlabel('Quality')
plt.ylabel('pH')
plt.title('Wine Data: Quality vs pH ')
```
